```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('dt_dataset2.csv',index_col="Employee ID ")
df
```

```python
plt.figure()
colors = {'A': 'blue', 'B': 'red'}
for t in df['Team'].unique():
    s = df[df['Team'] == t]
    plt.scatter(s['Gender'], s['Income'], c=colors[t], label=t)
plt.ylabel('Income')
plt.xlabel('Gender')
plt.legend()
plt.show()
```
